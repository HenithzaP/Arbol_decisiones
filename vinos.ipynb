```python
predictors_col = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH","sulphates", "alcohol"]
target_col = ['quality']
```

```python
predictors = data[predictors_col]
target = data[target_col]
```

Acidez vol√°til: Exhibe una correlaci√≥n positiva moderada con la acidez residual y el di√≥xido de carbono total, y una correlaci√≥n negativa moderada con el pH y la densidad.

√Åcido c√≠trico: Muestra una correlaci√≥n positiva moderada con las variables mencionadas anteriormente, y una correlaci√≥n negativa moderada con el pH y la densidad.

Az√∫car residual, cloruros y di√≥xido de azufre libre: Presentan patrones de correlaci√≥n similares a los mencionados anteriormente.

Estas correlaciones subrayan la complejidad de las interdependencias entre las caracter√≠sticas qu√≠micas del vino tinto, destacando la importancia de considerarlas en la produccion y evaluacion de la calidad del producto final.

**Mapa de Calor**

El mapa de calor proporciona una visualizaci√≥n clara de las relaciones entre las variables predictoras del vino tinto. Se observan correlaciones tanto positivas como negativas entre las variables, lo que proporciona una comprensi√≥n detallada de c√≥mo interact√∫an entre s√≠.
La acidez fija muestra una correlaci√≥n positiva moderada con la acidez residual y el di√≥xido de carbono total, mientras que presenta una correlaci√≥n negativa moderada con el pH y la densidad. De manera similar, la acidez vol√°til exhibe patrones de correlaci√≥n similares, lo que indica una relaci√≥n coherente con otras variables.
El √°cido c√≠trico, el az√∫car residual, los cloruros, el di√≥xido de azufre libre y el di√≥xido de azufre total tambi√©n muestran correlaciones significativas con las otras variables, tanto positivas como negativas, lo que sugiere interacciones complejas dentro del sistema del vino tinto.
La densidad, por otro lado, presenta una correlaci√≥n negativa moderada con todas las dem√°s variables, lo que indica una relaci√≥n inversa con los dem√°s componentes analizados.
En conclusi√≥n, el mapa de calor confirma la alta correlaci√≥n entre las variables predictoras del vino tinto, lo que sugiere una compleja interdependencia entre las variables analizadas.calidad.

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.2, random_state=13)
```

```python
tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit (X_train, y_train)
```

```python
plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)
```

```python
predicciones
```

```python
# Examinamos la matriz de confusi√≥n
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
accuracy = accuracy_score(y_test, predicciones)
accuracy
```

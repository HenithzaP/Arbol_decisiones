```python
predictors_col = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH","sulphates", "alcohol"]
target_col = ['quality']
```

```python
predictors = data[predictors_col]
target = data[target_col]
```

**Correlación**


La matriz de correlación revela relaciones significativas entre diversas variables predictoras del vino tinto.

Acidez fija: Posee una correlación positiva moderada con la acidez residual y el dióxido de carbono total, y una correlación negativa moderada con el pH y la densidad.

Acidez volátil: Exhibe una correlación positiva moderada con la acidez residual y el dióxido de carbono total, y una correlación negativa moderada con el pH y la densidad.

Ácido cítrico: Muestra una correlación positiva moderada con las variables mencionadas anteriormente, y una correlación negativa moderada con el pH y la densidad.

Azúcar residual, cloruros y dióxido de azufre libre: Presentan patrones de correlación similares a los mencionados anteriormente.

Estas correlaciones subrayan la complejidad de las interdependencias entre las características químicas del vino tinto, destacando la importancia de considerarlas en la produccion y evaluacion de la calidad del producto final.

```python
sns.heatmap(corr)
plt.show()
```

**Mapa de Calor**

El mapa de calor proporciona una visualización clara de las relaciones entre las variables predictoras del vino tinto. Se observan correlaciones tanto positivas como negativas entre las variables, lo que proporciona una comprensión detallada de cómo interactúan entre sí.
La acidez fija muestra una correlación positiva moderada con la acidez residual y el dióxido de carbono total, mientras que presenta una correlación negativa moderada con el pH y la densidad. De manera similar, la acidez volátil exhibe patrones de correlación similares, lo que indica una relación coherente con otras variables.
El ácido cítrico, el azúcar residual, los cloruros, el dióxido de azufre libre y el dióxido de azufre total también muestran correlaciones significativas con las otras variables, tanto positivas como negativas, lo que sugiere interacciones complejas dentro del sistema del vino tinto.
La densidad, por otro lado, presenta una correlación negativa moderada con todas las demás variables, lo que indica una relación inversa con los demás componentes analizados.
En conclusión, el mapa de calor confirma la alta correlación entre las variables predictoras del vino tinto, lo que sugiere una compleja interdependencia entre las variables analizadas.calidad.

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.2, random_state=13)
```

```python
tree = DecisionTreeClassifier()
```

```python
arbol = tree.fit (X_train, y_train)
```

```python
plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)
```

```python
predicciones
```

```python
# Examinamos la matriz de confusión
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
accuracy = accuracy_score(y_test, predicciones)
accuracy
```
